```python
# Importando modulos necesarios
import numpy as np
import matplotlib.pyplot as plt
import sympy 
from scipy import integrate


a = 0 # Limite inferior de la solucion en x
b = 2 # Limite superior de la solucion en y 
n = 100 # Numero de particiones 

h = (b-a)/n # Tamaño de paso

valores_x = [] # Vector donde iran los valores de x
valores_y = [] # Vector donde iran los valores de x

t = 0 # Condicion inicial
x = 1 # Condicion inicial

valores_x.append(t) # se van ajuntando los primeros valores de t y x
valores_y.append(x)

def funcion(t,x):  # Se define la funcion f(t,x)
    return 1 -t +4*x

for i in range(int(n)): # Se hace un ciclo for para obtener cada par de valores
    
    xn = x + h*funcion(t,x) # la funcion es igual a sus valores anteriores
    x = xn # El valor actual de x se vuelve xn
    t = t + h # Se aumenta un paso h 
    
    valores_x.append(t) # Se adjuntan los valores a los vectores t y x
    valores_y.append(x)
    

plt.plot(valores_x,valores_y) # Se grafica
 
plt.xlabel('t')
plt.ylabel('x') 
pl.savefig('Primera gráfica.png')
plt.show() 
```
